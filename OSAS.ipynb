# Data Preparation

```python
data = data.drop(columns=['PoolQC','MiscFeature', 'Alley', 'Fence'])

print(data.shape)
```

```python
def missingValues(data):
    num_cols = [cname for cname in data.columns if data[cname].dtype in ['int64', 'float64']]
    cat_cols = [cname for cname in data.columns if data[cname].dtype == "object"]
    values = {}
    for a in cat_cols:
        values[a] = 'UNKOWN'

    for a in num_cols:
        values[a] = data[a].median()
        
    data.fillna(value=values,inplace=True)
    
    
missingValues(data)
data.head()
```

```python
data.isnull().sum().sum()
```

```python
conditions = [
    (data['SalePrice'] <= 163000),
    (data['SalePrice'] > 163000)]

choices = ['Lower', 'Higher']

data['PriceRange'] = np.select(conditions, choices, default='Medium')

```

```python
d1 = data[data["PriceRange"] == "Lower"]
d2 = data[data["PriceRange"] == "Higher"]
```

```python
d1.shape
```

```python
d2.shape
```

```python
new_d = pd.concat([d1,d2])

new_d.shape
```

```python
def objectColList(data):
    return [cname for cname in data.columns if data[cname].dtype == "object"]

def doEncoding(data,columnsToEncode):
    return pd.get_dummies(data,columns = columnsToEncode)

cat_cols = objectColList(data)
data = doEncoding(data,cat_cols)
data.head()
```

# Data Mining

### Setup

```python
cols = data.columns.tolist()
cols = [c for c in cols if c not in ["SalePrice", "PriceRange_Higher", "PriceRange_Lower"]]
target = "SalePrice"
```

```python
from sklearn.model_selection import train_test_split

train = data.sample(frac=0.7, random_state=1)

test = data.loc[~data.index.isin(train.index)]

print(train.shape)
print(test.shape)
```

### Testing different models

```python
from sklearn.linear_model import RidgeCV

ridge_cv = RidgeCV(alphas=(0.01, 0.05, 0.1, 0.3, 1, 3, 5, 10))
ridge_cv.fit(train[cols], train[target])
```

```python
from sklearn.metrics import mean_squared_error

predictions = ridge_cv.predict(test[cols])

mean_squared_error(predictions, test[target])
```

```python
from warnings import simplefilter
simplefilter(action='ignore', category=FutureWarning)

from sklearn import svm

clf = svm.SVR(kernel='linear')
clf.fit(train[cols], train[target]) 
```

```python
from sklearn.metrics import mean_squared_error

predictions = clf.predict(test[cols])

mean_squared_error(predictions, test[target])
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(train[cols], train[target])
```

```python
from sklearn.metrics import mean_squared_error

predictions = model.predict(test[cols])

mean_squared_error(predictions, test[target])
```

### Final Model

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(train[cols], train[target])
```

```python
predictions = model.predict(test[cols])
```

# Interpretation

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error


print('Mean Absolute Error:', mean_absolute_error(predictions, test[target]))  
print('Mean Squared Error:', mean_squared_error(predictions, test[target]))  
print('Root Mean Squared Error:', np.sqrt(mean_squared_error(predictions, test[target])))
```
